```python
import random
import numpy as np
import pandas as pd

column_list = ['START(sec)', 'END(sec)', 'FLOWRATE']
tabular = pd.DataFrame(columns=column_list)

total_duration = 0
while True:
    flow_rate = random.randint(1, 250) * 5
    duration = random.randint(1, 1000) * 500
    temp_df = pd.DataFrame([[total_duration, total_duration + duration, flow_rate]], columns=column_list)
    tabular = tabular.append(temp_df, ignore_index=True)
    total_duration += duration
    if total_duration > 1.6 * 10**8:
        break
```

```python
tabular.to_excel('tabular.xlsx')
```
