```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time_array_onerate = pd.read_csv("../onerate/clean_onerate.csv", index_col=0)["Time(s)"].to_numpy()
outliers_onerate = pd.read_csv("../onerate/outliers_onerate.csv", index_col=0)["Pressure(psi)"].to_numpy()
time_array_bigdata = pd.read_csv("../bigdata/clean_bigdata.csv", index_col=0)["Time(s)"].to_numpy()
outliers_bigdata = pd.read_csv("../bigdata/outliers_bigdata.csv", index_col=0)["Pressure(psi)"].to_numpy()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
ax1.set_title("Pressure vs Time")
ax1.set_xlabel("Time(s)")
ax1.set_ylabel("Pressure(psi)")
ax2.set_title("Pressure vs Time")
ax2.set_xlabel("Time(s)")
ax2.set_ylabel("Pressure(psi)")
ax1.scatter(time_array_onerate, outliers_onerate, s=1)
ax2.scatter(time_array_bigdata, outliers_bigdata, s=1)
fig.savefig("outliers_onerate_and_bigdata", dpi=400, bbox_inches="tight")
```
